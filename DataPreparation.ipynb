```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DataPrepFinal.csv") #nomeando a nossa base como DF - abreviação de dataframe
df.head() #trazendo a vizualização da base
```

```python
df = df.drop(columns=['category','name', 'Unnamed: 0', 'pledged', 'usd pledged']) #a função drop é usada para limpar a base e excluir as colunas que não desejamos
```

```python
df.dtypes
```

```python
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD ')) #na coluna usd_goal_real há os caracteres USD, mas queremos que seja somente valor. Portanto, vamos excluir isso da coluna.
```

```python
df['launched'] = df['launched'].str.replace('\d\d:\d\d', '', regex=True) #na coluna launched, tem data e HORA, queremos remover a hora. Logo, usamos o REGEX para substituir esse campo por nada.
```

```python
df.head()
```

```python
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
```

```python
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y') #convertendo a coluna deadline em DATA
```

```python
df.head()
```

```python
df['launched'] = pd.to_datetime(df['launched'])
```

```python
df.head()
```

```python
df['time'] = (df['deadline']-df['launched']) #estou calculando a diferença de dias entre o lanched e o deadline, criando uma nova coluna chamada "time"
```

```python
df.head()
```

```python
df_invest = pd.read_csv('invested.csv')
```

```python
df_invest = pd.get_dummies(df_invest, columns= ['backedLocation']) #com esta medida ele vai ver quantos registros existem por localização
```

```python
df_invest.sort_values(by = ['ID']) # aqui vemos que a mesma ID possui mais de um registro, então não é uma dimensão
```
